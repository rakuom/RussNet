The change in population of each African country in the last 3 decades

```python
Data.head()
```

```python
Data['continent'].value_counts()
```

```python
Africa = Data[Data.continent=='Africa']
```

```python
Africa.head()
```

```python
Africa['continent'].value_counts()
```

```python
Afri_Data = Africa.groupby(['country','year'])['pop'].sum().reset_index(name='Population')
```

```python
Afri_Data = Afri_Data[Afri_Data['year']>1972]
```

```python
Afri_Data['year'].value_counts()
```

```python
Afri_Data.head()
```

```python
Data_Dec = Afri_Data[(Afri_Data.year == 1977)| (Afri_Data.year ==1987) | (Afri_Data.year==1997) | (Afri_Data.year==2007)]
```

```python
Data_Dec.head()
```

```python
Data_Dec.groupby(['country','year'])['Population'].sum().reset_index(name='Total Population')
```

```python
Population = pd.crosstab(Data_Dec.country, Data_Dec.year, values=Data_Dec.Population, aggfunc='sum')
```

```python
Population.head()
```

```python
Population = pd.DataFrame(Population)
```

```python
Population.head()
```

```python
Population['Decade_1'] = Population.apply(lambda row: row[1987] - row[1977], axis = 1)
```

```python
Population.head()
```

```python
Population['Decade_2'] = Population.apply(lambda row: row[1997] - row[1987], axis = 1)
```

```python
Population.head()
```

```python
Population['Decade_3'] = Population.apply(lambda row: row[2007] - row[1997], axis = 1)
```

```python
Population.head()
```

```python
New_Pop = Population[['Decade_1','Decade_2','Decade_3']]
```

```python
New_Pop.head()
```

```python
import matplotlib.pyplot as plt
def cm_to_inch(value):
    return value/2.54
plt.figure(figsize=(cm_to_inch(15),cm_to_inch(10)))
New_Pop.plot.bar()
plt.legend(title='Population Growth')

plt.show()
```
