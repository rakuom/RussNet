Demographic:
a. Which 2 demographics have proven to be the most popular with the telco brand?
b. If the marketing team wanted to increase customer retention, to whom should 
they target their advertisements to?

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head(50)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import plotly.express as px
```

```python
#Demographic data is gender,senior citizen,partner and dependents
df.head()
```

```python
Demographs = {'Gender':df.gender,
              'SeniorCitizen':df.SeniorCitizen,
              'Partner':df.Partner,
              'Dependents':df.Dependents,
              'Churn':df.Churn}
```

```python
df_Demo = pd.DataFrame(Demographs)
```

```python
df_Demo.head()
```

```python
df_Demo['Churn'].replace(to_replace='Yes',value=1,inplace=True)
df_Demo['Churn'].replace(to_replace='No',value=0,inplace=True)
```

```python
df_Demo.head(50)
```

```python
df_Dummies = pd.get_dummies(df_Demo)
```

```python
df_Dummies.head()
```

```python
df_Dummies.info()
```

```python
#Co-relation of churn with the demographics
plt.figure(figsize=(15,10))
df_Dummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar')
```

 Which 2 demographics have proven to be the most popular with the telco brand?
 From the co-relation plot, the 2 demographics most popular with the telco brand are:
 1. Non-Senior Citizens with partners
 2. Non-senior Citizens with dependents
 This is because this demographic has a negative co-relation with churn. They are not leaving the network.

If the marketing team wanted to increase customer retention, targetted advertisements should be to senior citizens without partners and senior citizens without dependents. This is because this demographic has a positive co-relation with churn. They are leaving the network.
